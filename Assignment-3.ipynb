#### Columns with skewness>0.5 or skewness<-0.5 are considered skewed. Columns: Runs scored,Hits,Home Runs,Complete Game,Shut Outs,Saves,Errors are skewed positively


```python
mean=data.mean()
std=data.std()

zscore=(data-mean)/std
```

```python
zscore.head()
```

```python
rows_to_drop1=(zscore<=-3 ).any(axis=1)
rows_to_drop2=(zscore>=3).any(axis=1)
df=data[~rows_to_drop1]
df2=data[~rows_to_drop2]
```
